```python
# 1. Which team has scored top 5 runs across all seasons?
top_team = ipl_df.groupby('batting_team')['runs_off_bat'].sum().reset_index()
top_scored_team = top_team.sort_values(by='runs_off_bat',ascending=False)
top5_team = top_scored_team.head(5)
plt.figure(figsize=(10,6))
plt.bar(top5_team['batting_team'], top5_team['runs_off_bat'],color="darkgreen")
plt.title('Top 5 Teams by Total Runs Scored Across All IPL Seasons',fontsize=14)
plt.xlabel("Batting Team",fontsize=14)
plt.ylabel('Runs off bat',fontsize=14)
plt.xticks(rotation=45)
plt.show
```

```python
# 2.  Which team has scored bottom 5 runs across all seasons?

import matplotlib.pyplot as plt
last5_team = ipl_df.groupby('batting_team')['runs_off_bat'].sum().reset_index()
last5_scored_team = last5_team.sort_values(by='runs_off_bat', ascending=True)
last_team = last5_scored_team.head(5)

# Assuming `top_scorring_team` DataFrame is already available
plt.figure(figsize=(10, 6))
plt.bar(last_team['batting_team'], last_team['runs_off_bat'], color='orange')
plt.xlabel('Total Runs', fontsize=14)
plt.ylabel('Teams', fontsize=14)
plt.title('Bottom 5 Teams by Total Runs across all IPL Seasons')

plt.xticks(rotation=45)
plt.show()
```

```python
# 3.  What is the top 5 number of extras conceded by each team?
top5_extras = ipl_df.groupby('bowling_team')['extras'].sum().reset_index()
top5_scored_extras = top5_extras.sort_values(by='extras',ascending=True)
top_extras = top5_scored_extras.tail(5)

plt.figure(figsize=(10,6))
plt.barh(top_extras['bowling_team'], top_extras['extras'], color='purple')
plt.title('Top 5 extras value wise team', fontsize=30)
plt.xlabel('Teams',fontsize=14)
plt.ylabel('Value of extras team', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
#4. What is the number of extras sepreated with wides, noballs, byes,legbyes,penalty conceded by each team?
divided_extras = ipl_df.groupby('bowling_team')[['wides','noballs','byes','legbyes','penalty']].sum()
extras_5 = divided_extras.sort_values(by=['wides','noballs','byes','legbyes','penalty'],ascending=False)


plt.figure(figsize=(10,6))
sns.heatmap(extras_5, annot=True, cmap='YlGnBu')
plt.title("Extras conceded by Each Team", fontsize=14)
plt.show()
```

```python
# 5. Who are the top 10 batsmen team with the most runs:

top_batsmen_teams = ipl_df.groupby(['striker', 'batting_team'])['runs_off_bat'].sum().reset_index()

top_batsmen_teams_sorted = top_batsmen_teams.sort_values(by='runs_off_bat', ascending=False).head(10)
plt.figure(figsize=(12,8))
plt.barh(top_batsmen_teams_sorted['striker'],top_batsmen_teams_sorted['runs_off_bat'],color="skyblue")
plt.xlabel("Runs")
plt.ylabel("batting_team")
plt.title("Top Batsmen with the Most Runs in International Cricket")
plt.gca().invert_yaxis()
plt.show()


```

```python
# 6. What are the distribution of wicket types?
wicket_distribution = ipl_df['wicket_type'].value_counts().reset_index()
wicket_distribution.columns = ['wicket_type', 'frequency']


plt.figure(figsize=(10, 6))
sns.countplot(x='wicket_type', data=wicket_distribution, palette='viridis')
plt.xlabel("Wicket Type")
plt.ylabel("Count")
plt.title("Distribution of Wicket Types")
plt.xticks(rotation=45)
plt.show()
```

```python

#7. year wise winners:
season_wise_wins = ipl_df.groupby(['season', 'batting_team'])['match_id'].nunique().reset_index()

# Find the team with the maximum wins in each season
season_wise_winners = season_wise_wins.loc[season_wise_wins.groupby('season')['match_id'].idxmax()]

# Rename columns for better understanding
season_wise_winners.columns = ['season', 'team', 'wins']

#print(season_wise_winners)

# Plotting with the correct DataFrame
plt.figure(figsize=(12, 8))

# Create a horizontal bar plot with the season-wise winners
plt.barh(season_wise_winners['season'], season_wise_winners['batting_team'], color="orange")

# Adding labels and title
plt.xlabel("Number of Wins")
plt.ylabel("Season")
plt.title("Season-wise Winners")

# Display the plot
plt.show()
```
