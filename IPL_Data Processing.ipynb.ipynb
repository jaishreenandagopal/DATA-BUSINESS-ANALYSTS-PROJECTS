```python
df.columns
```

```python
# 1. Which team has scored top 5 runs across all seasons?
total_runs_by_teams = ipl_df.groupby("batting_team")['runs_off_bat'].sum().reset_index()
top_scorring_team = total_runs_by_teams.sort_values(by="runs_off_bat", ascending=False).head(5)
top_scorring_team
```

```python
# 2. Which team has scored bottom 5 runs across all seasons?
total_runs_by_teams = ipl_df.groupby("batting_team")['runs_off_bat'].sum().reset_index()
top_scorring_team = total_runs_by_teams.sort_values(by="runs_off_bat", ascending=False).tail(5)
top_scorring_team
```

```python
# 3. What is the top 5 number of extras conceded by each team?
average_extras_by_teams = ipl_df.groupby('bowling_team')['extras'].sum().reset_index()
average_extras_by_teams_sum = average_extras_by_teams.sort_values(by="extras",ascending=False).head()
average_extras_by_teams_sum
```

```python
#4. What is the number of extras sepreated with wides, noballs, byes,legbyes,penalty conceded by each team?
average_extras_by_teams = ipl_df.groupby('batting_team')[['wides','noballs','byes','legbyes','penalty']].sum().reset_index()
average_extras_by_teams.sort_values(by=['wides','noballs','byes','legbyes','penalty'],ascending=False,inplace=True)
average_extras_by_teams
```

```python
# 5. Who are the top 10 batsmen team with the most runs:

top_batsmen_teams = df.groupby(['striker', 'batting_team'])['runs_off_bat'].sum().reset_index()

top_batsmen_teams_sorted = top_batsmen_teams.sort_values(by='runs_off_bat', ascending=False).head(10)

print(top_batsmen_teams_sorted)
```

```python
# 6. What are the distribution of wicket types?
wicket_distribution = ipl_df['wicket_type'].value_counts().reset_index()
wicket_distribution.columns = ['wicket_type', 'frequency']
print(wicket_distribution)
```

```python
#7. year wise winners:
season_wise_wins = df.groupby(['season', 'batting_team'])['match_id'].nunique().reset_index()

# Find the team with the maximum wins in each season
season_wise_winners = season_wise_wins.loc[season_wise_wins.groupby('season')['match_id'].idxmax()]

# Rename columns for better understanding
season_wise_winners.columns = ['season', 'team', 'wins']

print(season_wise_winners)
```
